```python
import glob
import os
from pathlib import Path
```

```python
os.getcwd()
```

```python
from PIL import Image

images_path = Path('../../images')
s = 0
shapes = set()
for folder in os.listdir(images_path):
    # print(f"{folder}: {len(os.listdir(images_path / folder))}")
    for image_files in os.listdir(images_path / folder): 
        img = Image.open(images_path / folder / image_files)
        shapes.add(img.size)

    s += len(os.listdir(images_path / folder))
print(f"TOTAL: {s}")
print(shapes)
```
